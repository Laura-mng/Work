```python
#5)Italia,Spagna,Francia,Germania nel 2023,pazienti ospitalizzati per ognuna.Gestire i dati nulli.
```

```python
data3=(df["date"]>="2023-01-01")&(df["date"]<="2023-12-31")
```

```python
paesi3=df["location"].isin(["Italy","Germany","France","Spain"])
data_paese3=df.loc[paesi3&data3]
data_paese3
```

```python
pazienti_osp=data_paese3.groupby("location")["hosp_patients"].sum()
pazienti_osp.reset_index()
```

```python
#T dati nulli possiamo gestirli in due modi: o eliminandoli con .dprona(), oppure li sostituiamo con dei paramentri significativi con la funzione .str.replace();
#In questo caso i singoli dati,posti data per data, possono avere un significato ben preciso per capire l'andamento del paese e dei casi,ed avere un'analisi più accurata ed approfondita.
```

```python
#N.B: Dal momento che il Germania,nel 2023 il numero dei casi ospitalizzati è 0, ha un significato ben preciso: 
#Lo zero in questione mi identifica che in quel determinato momento,in tale paese non ci sono stati dei casi ospitalizzati.
#Questo quantifica l'efficacia delle misure di cautela e di organizzazione del paese.
```
